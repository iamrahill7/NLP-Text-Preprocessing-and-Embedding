# Word Embedding

* Word embedding is a technique in Natural Language Processing (NLP) where words are represented as numerical vectors in a high-dimensional space.

* These vectors capture the semantic relationships between words, meaning similar words will have similar vector representations.

* Word representations are broadly classified into **count-based (traditional)** and **deep learning-based** methods.  

---

## **1️⃣ Count-Based Methods** (Traditional NLP)  
- **Bag of Words (BoW)**  
- **One-Hot Encoding (OHE)**  
- **Term Frequency-Inverse Document Frequency (TF-IDF)**  

These methods rely on word frequencies but lack semantic understanding.  

---

---

### Why TF-IDF and BoW Are Not Ideal for Semantic Search  

1. **Lack of Semantic Understanding** – TF-IDF (Term Frequency-Inverse Document Frequency) and BoW (Bag of Words) only count word occurrences without understanding their meaning or context. They treat words independently, which limits their ability to capture relationships between words.  

2. **Sparse Matrices & Slow Processing** – These methods generate large sparse matrices, making computations slow and memory-intensive.  

### 🧠 Why Word2Vec is a Better Alternative  

---

## **2️⃣ Deep Learning-Based Word Embeddings**  

### **Word2Vec**  

* Word2Vec is a neural network-based technique developed by Google in 2013 for learning word embeddings.

**Advantages of Word2Vec over TF-IDF and BoW:**  
- **Captures Meaning** – Words with similar meanings have similar vectors, improving search accuracy.  
- **Context Awareness** – Word2Vec learns word relationships based on surrounding words, unlike TF-IDF and BoW.  
- **Less Sparse & Faster** – It produces compact, dense vectors, making computations more efficient.  
- **Handles Synonyms** – Words with similar meanings are placed close in vector space, improving NLP tasks.  

### 🥜 In a nutshell
- TF-IDF and BoW are useful for basic text analysis but fail in semantic search due to their lack of contextual understanding and inefficiency.
- Word2Vec, on the other hand, learns relationships between words, making it far superior for tasks requiring meaning-based search and NLP applications.

```python
vector = word2vec["king"]
print(vector)  # 300-dimensional array
```

### **What Do the 300 Numbers Mean?**
* Each number in the vector tells how much the word relates to different hidden meanings.
* For example, for "king", some dimensions might represent:

1. **Gender** (more male)
2. **Royalty** (high score)
3. **Leadership** (high score)
4. **Plurality** (low score, because "king" is singular)
... and so on for 300 different properties.
* That's why "king" has 300 numbers! Each number tells how strong or weak "king" is in different aspects.

```python
vector.shape
```

The **300 numbers represent how "king" behaves in different scenarios** (gender, royalty, leadership, etc.).  

### **Will Every Word Have 300 Numbers?**  
✅ **Yes!** If you use a **Word2Vec model trained with 300 dimensions**, every word will have **exactly 300 numbers**, no matter if it's "king", "apple", or "universe".  

### **What If a Word Has More Than 300 Vectors?**  
That depends on **the model**.  

🔹 **In Word2Vec, the number of dimensions is fixed** when the model is trained.  
🔹 If a model is trained with **400 dimensions**, then every word will have **400 numbers** instead of 300.  
🔹 If trained with **200 dimensions**, every word will have only **200 numbers**.  

💡 **So, the number of dimensions (300, 400, etc.) is decided when training the model!**

### **Let's look at similar words**

```python
similar_words = word2vec.most_similar("king", topn=10)
similar_words
```

### **Cosine Similarity**

```python
similarity = word2vec.similarity("king", "queen")
print(similarity)  # Higher value means more similarity
```

```python
similarity = word2vec.similarity("salary", "stipend")
print(similarity)  # Higher value means more similarity
```

### **Let's visualize the same word by using a Word Cloud**

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
word = "king"  # Change this to any word you want
similar_words = word2vec.most_similar(word, topn=20)  # Get top 20 similar words

# Convert to dictionary {word: similarity score}
word_dict = {w: s for w, s in similar_words}
```

```python
wordcloud = WordCloud(width=800, height=400, background_color="black").generate_from_frequencies(word_dict)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title(f"Word Cloud of Words Similar to '{word}'")
plt.show()
```
